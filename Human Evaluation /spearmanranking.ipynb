```python

def calculate_leave_one_out_spearman_detailed(scores):

    # renamed variables to be clear to all readers
    """
    Calculates "leave-one-out" Spearman's rank correlation coefficients

    Parameters:
    - scores: A 3D numpy array of shape (c, p, q):
              c is the number of conversations,
              p is the number of people (including GPT),
              q is the number of questions.

    Returns:
    - individual_correlations: A dictionary with keys as comparisons and values as Spearman's correlation coefficients.
    - average_correlation: The average of all individual Spearman's rank correlation coefficients.
    """

    c, p, q = scores.shape
    individual_correlations = {}
    correlations = []
    averages_by_person_left_out = {}

    for person_left_out in range(p):
        person_correlations = []
        for conv_index in range(c):
            indices_to_include = [i for i in range(p) if i != person_left_out]
            for person_to_compare in indices_to_include:
                # calculate Spearman's rank correlation for each pair
                rho, _ = spearmanr(scores[conv_index, person_left_out, :], scores[conv_index, person_to_compare, :])
                # store each correlation, each key is clearly described
                key = f"\nConv {conv_index+1}, Person {person_left_out+1} vs. Person {person_to_compare+1}"
                individual_correlations[key] = rho
                correlations.append(rho)
                person_correlations.append(rho)

        # calculate and store average correlation for the current person left out
        person_average = np.mean(person_correlations)
        averages_by_person_left_out[f'Person {person_left_out+1}'] = person_average
    # not very meaningful but we calculate the overall average of the correlations
    overall_average_correlation = np.mean(correlations)

    # Printing results
    print("{")
    print("  'individual_correlations': {")
    for key, value in individual_correlations.items():
        print(f"    '{key}': {value},")
    print("  },")
    print("  'averages_by_person_left_out': {")
    for key, value in averages_by_person_left_out.items():
        print(f"    '{key}': {value:.3f},")
    print("  },")
    print(f"  'overall_average_correlation': {overall_average_correlation:.2f}")
    print("}")

    return individual_correlations, averages_by_person_left_out, overall_average_correlation
```
